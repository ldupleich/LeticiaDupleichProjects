# Experimenting with LightGBM

In order to understand how to use the LightGBM model, let us download it and experiment a little with it. For the majority of this notebook, I will be using the information from their official website to compose a trial code.

LightGBM website: https://lightgbm.readthedocs.io/en/stable/

```python
import lightgbm as lgb
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
```

### Defining the dataset

For experimenting with LightGBM, lets use a dataset that we know requires little preprocessing, contains mixed datatypes (as we want to test that feature of LightGBM for our own blockbuster model), and whose patterns are already known so it is simple to test the accuracy.

For this, I have used the Titanic dataset, a famous dataset that tends to be used for classification models, specifically to determine if a passenger with specific features was likely to survive or not. 

Here is a summary of the dataset (this was directly copied from the notebook from my previous class):

- *PassengerId*: passenger ID assigned in this dataset
- *Survived*: a Boolean indicating whether the passenger survived or not (0 = No; 1 = Yes); this is our target
- *Pclass*: passenger class (1 = 1st; 2 = 2nd; 3 = 3rd)
- *Name*: field containing the name and title of the passenger
- *Sex*: male/female
- *Age*: age of the passenger in years
- *SibSp*: number of siblings/spouses aboard
- *Parch*: number of parents/children aboard
- *Ticket*: ticket number
- *Fare*: passenger fare in British Pounds
- *Cabin*: location of the cabin, consisting of a letter indicating the deck, and a cabin number.
- *Embarked*: port of embarkation (C = Cherbourg; Q = Queenstown; S = Southampton)

```python
data = pd.read_csv('titanic.csv')
```

### Preprocessing

Thankfully, in a previous class, I have used this dataset so I know that cleaning it is relatively simple. I followed the same steps as I did for my previous class

1. Display what the dataset looks like and look for any missing values or useless columns.
2. Remove the columns that will not be necessary for the analysis
3. If the column has very few missing values, just drop those entries
4. If the column has many missing values, find a way to fill them (e.g. using the average for "Age" column)

```python
data.info()
```

At first sight, we can immediately notice that there are some features that will not be useful for our model since they do not directly impact whether a passenger will survive or not. These features are: 

- PassengerID
- Name
- Ticket
- Cabin

There are also some emptyentries in the "Age" and "Embarked" columns. Since "Embarked" only has two missing entries, we can just drop these. "Age" has many more missing entries, so we can just fill these by calculating the mean of the ages of all the passengers.

```python
# Removing the columns that I will not be using
data = data.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)

# Drop rows where "Embarked" is missing
data = data.dropna(subset=['Embarked'])

# Adding proxy values to the empty entries in "Age"
mean = data["Age"].mean()
data = data.fillna({"Age": mean})

data.info()
```

### Features and variables

According to the LightGBM website, we first need to let the model know what our categorical features are so that it knows how to handle them accordingly. By looking at the data's info, we can see that the "Sex" and "Embarked" columns are categorical.

```python
categorical_columns = ["Sex", "Embarked"]
label_encoders = {}

for col in categorical_columns:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    label_encoders[col] = le
```

Now we have to define the feature variables and the target variables. Since I want to be able to predict whether a passenger has surivived, that is our target variables. So, in order to obtain our features (x), we have to drop that variable from our dataset and save it as our target in y.

```python
# We will use x as the feature
x = data.drop("Survived", axis=1)

# We will use y as the target
y = data["Survived"]
```

Now, lets split our data into training and testing so that we can start using the model with the training set, and then determine the accuracy of our findings with our testing set. First we have to split the data (the common split is 70% training and 30% testing), and then we have to give LightGBM each of the datasets.

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=40)
```

```python
train_data = lgb.Dataset(
    X_train,
    label=y_train,
    categorical_feature=categorical_columns,
    free_raw_data=False
)

test_data = lgb.Dataset(
    X_test,
    label=y_test,
    categorical_feature=categorical_columns,
    reference=train_data,
    free_raw_data=False
)
```

### Parameters

On the LightGBM website there are many parameters, so understanding all of them would take a lot of time. In order to facilitate this process and determine which parameters are the most important to specify every time I want to use the model, I asked ChatGPT to give me an overview of the most important ones to keep in mind. Below I am attaching the prompts I gave it and the replies that were provided.

ChatGPT parameter conversation: https://chatgpt.com/share/684fff01-80c4-8003-a8a6-178d8e54f148

---

<u>Core Parameters</u>

**objective:** Defines what type of model we want to use for the loss function. For this test model we want to make use of binary class since a passenger can either survive (1) or not (0). 
- Possible values: regression, regression_l1, huber, fair, poisson, quantile, mape, gamma, tweedie, binary, multiclass, multiclassova, cross_entropy, cross_entropy_lambda, lambdarank, rank_xendcg, aliases: objective_type, app, application, loss

**boosting_type:** The type of boosting algorithm that we want to use. Since we want to use gradient boosting decision trees (gbdt) for our blockbuster model, I will be attempting to use that one in this notebook as well.
- Possible values: gbdt, rf, dart, aliases: boosting_type, boost

**num_leaves:** Defines the total number of leaves that will be present in the tree. Here is where we have to be careful about overfitting if we have too many number of leaves.
- The default value for this is 31

**learning_rate:** The step size that we want to take after every iteration of the algorithm. If we are using trees, this is the contribution that each tree has to the final, so a smaller alpha value requires more trees.
- The default value for this is 0.1

---

<u>Learning Parameters</u>

**feature_fraction:** The fraction of features that we will be using for building each tree. Again, this is another place where we have to be careful with overfitting.
- Typical values are between 0.6 and 1

**bagging_fraction:** The fraction of features that will be used for each iteration (not for each tree like feature_fraction).
- The default value for this is 1.0

**bagging_freq:** How many 'bags' will be created. Creating bags means that we will train different trees (in our case) and then our final one is a combination of the predictions.
- The default value for this is 0

**verbose:** Controls the messages that will be printed. So either no debug info or extensive debug information.
- The default value for this is 1

---

<u>Metric Parameters</u>

**metric:** Defines how the model's predictions are evaluated. For this test model we want to make use of binary_logloss. 
- There are many possible values, please reference the LightGBM parameter website and go to the "Metric Parameters" section for more information.

---

Using this information, we can now define the parameters that will be used for this specific model.


```python
base_params = {
    'objective': 'binary',
    'num_class': 1,
    'metric': 'binary_logloss',
    'boosting_type': 'gbdt',
    'num_leaves': 100,
    'learning_rate': 0.1,
    'feature_fraction': 0.9,
    'bagging_fraction': 1,
    'bagging_freq': 5,
    'verbose': -1,
    'early_stopping_round': 50
}
```

With these parameters, we can now define the model that we will use to classify the products purchased. The "valid_sets" is another built in parameter of LightGBM which allows us to separate the data that we want to use during validation.

```python
model = lgb.train(
    base_params,
    train_data,
    valid_sets=[test_data],
    num_boost_round=500,
)
```

Since we now have a model, we can start making the predictions and defining what the classes for each prediction are, so that we can later calculate the accuracy of our model. For this we will be using the built in parameter "booster.predict" that can be found here https://lightgbm.readthedocs.io/en/stable/pythonapi/lightgbm.Booster.html.

```python
# Make predictions
y_pred = model.predict(X_test)
y_pred_binary = [1 if pred > 0.5 else 0 for pred in y_pred]
```

Finally, we can print the accuracy score of our model.

```python
# Calculate accuracy and print a classification report
print("Accuracy:", accuracy_score(y_test, y_pred_binary))
print(classification_report(y_test, y_pred_binary))
```

### Parameter Tuning and K fold Validation

Since we have an initial prediction accuracy, we can now focus on enhancing our model. In order to do this, we can do k fold validation and parameters tuning all in one to determine the best set of parameters. Luckily, LightGBM also has a cross validation section: https://lightgbm.readthedocs.io/en/stable/pythonapi/lightgbm.cv.html.

```python
eval_results = lgb.cv(
    base_params,
    train_data,
    nfold=5,
    num_boost_round=200,
    stratified=True,
    seed=40,
    eval_train_metric=True
)

print(eval_results.keys())
```

Now we can calculate the training and validation cost of our current model using this cross validation parameters.

```python
# Get training and validation loss at the final boosting round
# This was done using the names from the dict_keys above
avg_train_loss = eval_results['train binary_logloss-mean'][-1]
avg_val_loss = eval_results['valid binary_logloss-mean'][-1]

# Print average costs
print("Average training cost (logloss):", avg_train_loss)
print("Average validation cost (logloss):", avg_val_loss)
```

In order to see if our model is working properly, and not overfitting, it is important to see the learning curve of the model and ensure that it has the right shape. If the training cost is decreasing, but the validation cost is increasing, then that is probably an indication that the model is overfitting. If this was the case, we would have to find ways to decrease the complexity of the model, like removing leaves or setting a tree height limit. 

Lets run the code below and see how the learning curves look like to determine our next steps.

```python
plt.plot(eval_results['train binary_logloss-mean'], label='Train Loss')
plt.plot(eval_results['valid binary_logloss-mean'], label='Validation Loss')
plt.xlabel('Boosting Round')
plt.ylabel('Log Loss')
plt.title('LightGBM Training vs Validation Loss')
plt.legend()
plt.show()
```

Since it appears like the model is not overfitting, and the cost is decreasing per boosting round, we can now try a little hyperparameter tuning just to see if we can get even nice learning curves. However, this is already a pretty nice result. 

Now we can move onto parameter tuning to find the best parameters and the minimal cost. This will look a little repetitive as we just did the cross validation, but we will need to do more cross validations to find the best parameters. The first step is making a grid with all the different parameter values that we want to test, and therefore optimize. When we look at the website of LightGBM under the section "Parameter Tuning", they provide us with some examples of the parameters that we should focus on changing: https://lightgbm.readthedocs.io/en/stable/Parameters-Tuning.html#tune-parameters-for-the-leaf-wise-best-first-tree.

- num_leaves
- min_data_in_leaf

```python
# This tool allows us to test every possible combination of parameters
# This was obtained by asking ChatGPT: How can I test every combination of parameters in a Python code?
from itertools import product

# The default for num_leaves is 31, and the default for min_data_in_leaf is 20 so lets 
# look at values that are below and above that default
leaf_grid = {
    'num_leaves': [5, 10, 15, 31, 60, 120],
    'min_data_in_leaf': [2, 5, 10, 20, 50]
}

# These are what we want to optimize (loss and parameters)
best_cost = float('inf')
best_params = None

results = []

# We will be testing for all of the different parameter combinations
for num_leaves, min_data in product(leaf_grid['num_leaves'], leaf_grid['min_data_in_leaf']):
    params = base_params.copy()
    params['num_leaves'] = num_leaves
    params['min_data_in_leaf'] = min_data

    # For each combination we will get the validation cost (val_logloss)
    cv_results = lgb.cv(
        params,
        train_data,
        nfold=5,
        num_boost_round=200,
        stratified=True,
        seed=40,
        eval_train_metric=True,
    )

    val_logloss = cv_results['valid binary_logloss-mean'][-1]
    print(val_logloss)
    
    # If the validation cost of this combination is less than that of the best one, update it
    if val_logloss < best_cost:
        best_cost = val_logloss
        best_params = {
            'num_leaves': num_leaves,
            'min_data_in_leaf': min_data
        }
    
    results.append((num_leaves, min_data, val_logloss))


# Print the best parameter combination and the loss for that combination
print("\nBest parameter combination: ", best_params)
print("Loss for this combination: ", best_cost)
```

### Visualizing Cost

Although this step might not be extremely useful since we already have a log of how the cost changes depending on the combinations, it is often helpful to see how the different combinations affect the cost visually. For this reason, I asked ChatGPT what are the best ways to visualize hyperparameter tuning.

For the sake of exploration, I will be attempting to create all three plots that ChatGPT suggested:

1. Heatmap
2. Contour plot
3. 3D surface plot

Lets start by creating a heatmap using the seaborn library.

```python
import seaborn as sns

# Convert results to DataFrame
df = pd.DataFrame(results, columns=['num_leaves', 'min_data_in_leaf', 'val_logloss'])

# Pivot for heatmap
pivot = df.pivot(index='num_leaves', columns='min_data_in_leaf', values='val_logloss')

# Plot
plt.figure(figsize=(10, 6))
sns.heatmap(pivot, annot=True, fmt=".3f", cmap="viridis")
plt.title("Cost vs num_leaves and min_leaf")
plt.xlabel("num_leaves")
plt.ylabel("min_data_in_leaf")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

# Convert results to numpy arrays
leaves = sorted(df['num_leaves'].unique())
min_data = sorted(df['min_data_in_leaf'].unique())
X, Y = np.meshgrid(min_data, leaves)
Z = df.pivot(index='num_leaves', columns='min_data_in_leaf', values='val_logloss').values

# 3D plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax.set_xlabel('num_leaves')
ax.set_ylabel('min_data_in_leaf')
ax.set_zlabel('cost')
plt.title('3D Surface Plot of Cost')
plt.show()
```
